```python
#Alejandro Morales Madrigal

#Monte Carlo simulation of arithmetic Brownian motion
#dX(t) = mu*dt + sigma*dW(t)

import numpy as np
import matplotlib.pyplot as plt
```

```python
npaths = 2000;
T = 1;                         #time horizon
nsteps = 200;                  #number of time steps
dt = T/nsteps;                 #time step
t=np.linspace(0,T,num=nsteps); #observation times
mu = -0.05; 
sigma = 0.4;                   #model parameters
```

```python
#Monte Carlo
#Compute the increments with Euler-Maruyama
dX = mu*dt + sigma*np.random.normal(0,1,(npaths,nsteps))*np.sqrt(dt);
X = np.cumsum(dX,axis=1);
EX = mu*t; #Expected Path
```

```python
#Compute Variance
Xvar = np.ones([nsteps,1])
for j in range(nsteps):
    Xvar[j] = np.var(X[:,j])
```

```python
#Sample Paths
plot0 = plt.figure(0,figsize=(8,4), dpi=100)
plt.title('Arithmetic Brownian motion dX(t) = mu*dt + sigma*dW(t))')
for i in range(30):
    plt.plot(t,X[i+40,:])
plt.plot(t,EX,'black') #Expected Path
plt.tight_layout()
plt.xlim([0,1])
plt.ylim([0-1,1])
plt.xlabel('t')
plt.ylabel('X')
plt.grid()
plt.show()

plot1 = plt.figure(1,figsize=(6,4), dpi=100)
plt.title('Variance')
plt.plot(t,t*sigma**2) #Theoretical Variance
plt.plot(t,Xvar)       #Empirical Variance
plt.xlabel('t')
plt.ylabel('Var(X) = E((X-E(X))^2)')


plot2 = plt.figure(2,figsize=(8,6), dpi=100)
x=np.linspace(-1,1,num = 60);
plt.subplot(3,1,1)
plt.title('Density Functions at diferent times')
plt.hist(X[:,30],x)
plt.tight_layout()

plt.subplot(3,1,2)
plt.hist(X[:,90],x)
plt.tight_layout()

plt.subplot(3,1,3)
plt.hist(X[:,199],x)
plt.tight_layout()
```
