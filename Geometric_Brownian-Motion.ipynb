```python
#Monte Carlo simulation of geometric Brownian motion
#dS = mu*S*dt + sigma*S*dW

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define parameters and time grid
npaths = 20000 # number of paths
T = 1 # time horizon
nsteps = 200 # number of time steps
dt = T/nsteps # time step
t=np.linspace(0,T,num=nsteps) # observation times
mu = .5 
sigma = 0.4
S0 = 1 # initial stock price
```

```python
# Monte Carlo

# Compute the increments of the arithmetic Brownian motion X = log(S/S0)
dX = (mu-0.5*(sigma**2))*dt + sigma*np.random.normal(0,1,(npaths,nsteps))*np.sqrt(dt);

#Â Accumulate the increments
X = np.cumsum(dX,axis=1);

# Transform to geometric Brownian motion
S = S0*np.exp(X) #podriamos simular tambien con euler maruyama pero no lo hicimos aqui

# Expected Path
EX = np.exp(mu*t)
```

```python
#Sample Paths
plot0 = plt.figure(0,figsize=(7,4), dpi=100)
plt.title('Geometric Brownian motion dS = mu*S*dt + sigma*S*dW')
for i in range(60):
    plt.plot(t,S[i+40,:])
plt.plot(t,EX,'black') #Expected Path
plt.tight_layout()
plt.xlim([0,1])
plt.ylim([0,4])
plt.xlabel('t')
plt.ylabel('X')
plt.grid()
plt.show()

plot2 = plt.figure(2,figsize=(8,6), dpi=100)
x=np.linspace(0,3,num = 60);
plt.subplot(3,1,1)
plt.title('Density Functions at diferent times')
plt.hist(S[:,30],x)
plt.tight_layout()

plt.subplot(3,1,2)
plt.hist(S[:,90],x)
plt.tight_layout()

plt.subplot(3,1,3)
plt.hist(S[:,199],x)
plt.tight_layout()
```
